2. Real Placement Perspective
Suppose you have an array of 10 lakh elements.
Question:
Find whether number 999999 exists.
You can do it in two ways:

Method 1 — Linear Search
for(i = 0; i < n; i++)
{
    if(arr[i] == target)
        return i;
}

Worst case:
You check every element.
Operations ≈ n
Big-O = O(n)

Method 2 — Binary Search
while(low <= high)
{
    mid = (low + high)/2;

    if(arr[mid] == target)
        return mid;

    else if(arr[mid] < target)
        low = mid + 1;

    else
        high = mid - 1;
}

Each step cuts search space into half.
Big-O = O(log n)

For:
n = 1,000,000
Linear search may need:
10,00,000 comparisons
Binary search needs only about:
20 comparisons
That is why companies care about Big-O.

3. Why “O” ?
“O” means Order of Growth.
It focuses on:
trend
scalability
growth rate
Not exact execution time.
Because execution time depends on:
processor
RAM
compiler
language
Big-O ignores all these and studies growth mathematically.

8. How to Calculate Big-O Quickly

Rule 1 — Drop Constants
for(i = 0; i < 2*n; i++)

Complexity:
O(2n)
But Big-O ignores constants.
Final:
O(n)

Rule 2 — Keep Fastest Growing Term
n² + n + 10

Dominant term:
n²
Final:
O(n²)
Because when n becomes huge:
n² dominates everything.

9. Common Patterns

Single Loop

→ O(n)

Nested Loop
for(i=0; i<n; i++)
    for(j=0; j<n; j++)

→ O(n²)

Loop Dividing by 2
while(n > 1)
{
    n = n / 2;
}

→ O(log n)

Two Separate Loops
for(i=0; i<n; i++) {}
for(j=0; j<n; j++) {}

→ O(n + n)
→ O(n)
NOT O(n²)
This is a common placement mistake.

Big O Notation is how we describe the efficiency of an algorithm — not in seconds, but in terms of how the work grows as the input size grows. Companies like Google, Amazon, and Microsoft test this heavily in interviews because writing correct code isn't enough. You need to write scalable code.
The core question is: "If I double the input size, what happens to the time/space my algorithm needs?"

Here's a visual map of all the major complexities you need to know:---
Now let's go through each complexity with a concrete example — the kind an interviewer will actually ask you:
O(1) — Constant array[0] — Accessing an element by index. Doesn't matter if the array has 10 or 10 million elements. One operation, always.
O(log n) — Logarithmic Binary search. Every step cuts the problem in half. Searching 1 billion elements? Only ~30 comparisons. This is gold in interviews.
O(n) — Linear A single loop through an array. You touch every element once. Very acceptable.
O(n log n) — Linearithmic The best you can do for comparison-based sorting — Merge Sort, Heap Sort, Quick Sort (average). This is the sweet spot for sorting.
O(n²) — Quadratic A nested loop. Bubble Sort, Selection Sort, Insertion Sort. Fine for n=100, painful for n=10,000, a disaster for n=1,000,000.
O(2ⁿ) — Exponential Solving the Fibonacci sequence naively with recursion. Every new element doubles the work. Never acceptable at scale.
O(n!) — Factorial Brute-force permutations. n=12 already means 479 million operations. Avoid completely.

The three rules interviewers expect you to know:
Drop constants — O(2n) is just O(n). We care about the shape of growth, not the multiplier.


Drop non-dominant terms — O(n² + n) is just O(n²). As n gets large, the smaller term becomes irrelevant.


Worst case is the default — Unless told otherwise, Big O means worst-case. Binary search is O(log n) even though it could find the element on the first try.
